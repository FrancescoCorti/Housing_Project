```python
import pandas as pd
```

Unemployment (province) (2014-2024)

```python
df = pd.read_csv('C:/Users/HP/Desktop/Traineeship/data/unemployment_province.csv')
```

Unpivot the data

```python
unpivoted = df.melt(
    id_vars="year",
    var_name="prov_name",
    value_name="unemployment_prov"
)
```

Load dataset with updated prov codes

```python
name = pd.read_csv('datasets/prov_istat_mapping.csv')
```

Merge new prov name

```python
df1 = pd.merge(unpivoted, name, on = 'prov_name', how = 'inner')
```

Sobstitute Naples code to NAP to avoid problems with null-values

```python
df1.loc[df1['prov_name'] == 'Napoli','prov'] = 'NAP'
```

Save the data

```python
df1.to_parquet('datasets/macro/prov_unemployment.parquet', index=None)
```
