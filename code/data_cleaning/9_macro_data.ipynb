```python
import pandas as pd
```

Birth rate (lagged 2013-2023) (https://fred.stlouisfed.org/series/SPDYNCBRTINITA)

```python
df_birth = pd.read_csv('C:/Users/HP/Desktop/Traineeship/data/FRED/birth_rate_FRED.csv', index_col=False)

df_birth['observation_date'] = pd.to_datetime(df_birth['observation_date'])

df_birth['year'] = pd.DatetimeIndex(df_birth['observation_date']).year

df_birth = df_birth.loc[df_birth['year'] >= 2013].rename(columns = {'SPDYNCBRTINITA' : 'birth_rate'}).drop(columns = ['observation_date'])

df_birth.to_parquet('datasets/macro/lag_birth_rate.parquet', index=False)
```

Consumer price index (lagged) (https://fred.stlouisfed.org/series/ITACPIALLMINMEI)

```python
df_cpi = pd.read_csv('C:/Users/HP/Desktop/Traineeship/data/FRED/consumer_price_index_FRED.csv', index_col=False)

df_cpi['observation_date'] = pd.to_datetime(df_cpi['observation_date'])

df_cpi['year'] = pd.DatetimeIndex(df_cpi['observation_date']).year

df_cpi = df_cpi.loc[(df_cpi['year'] >= 2013) & (df_cpi['year'] <= 2023)].rename(columns = {'ITACPIALLMINMEI' : 'cpi'}).drop(columns = ['observation_date'])

df_cpi = df_cpi.groupby(['year'], as_index = False).agg({'cpi' : 'mean'})

df_cpi.to_parquet('datasets/macro/lag_cpi.parquet', index=False)
```
