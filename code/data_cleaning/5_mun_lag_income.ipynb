```python
# keep only the year
df_in['date'] = pd.to_datetime(df_in['date'])
df_in['year'] = df_in['date'].dt.year

# add missing zeroes to istat codes
add_zeroes(df_in, 'mun_istat', 6)
```

```python
new_order = [
    'mun_istat', 'year', 'class', 'income'
]

df_in = df_in[new_order]
```

Merge the datasets and create a new variable for income

```python
df_merged = pd.merge(df, df_in, on = ['mun_istat', 'year', 'class'], how = 'left')
```

```python
# drop rows with 0 and NaN values in taxpayers
df_nona = df_merged[df_merged['taxpayers'].notna() & (df_merged['taxpayers'] != 0)]

# create a variable class_mun_gdp as the class income (sum of incomes in a class / taxpayers in that class)
df_nona['class_mun_gdp'] = df_nona['income'] / df_nona['taxpayers']
```

```python
# check unique values in class
df_nona['class'].unique()
```

```python
# delete E_UN0 (values less or equal to one)
df_nona = df_nona[df_nona['class'] != ('E_UN0')]

# rename values in class
df_nona['class_mid'] = df_nona['class'].replace({
    'E0-10000': 5000,
    'E10000-15000': 12500,
    'E15000-26000': 20500,
    'E26000-55000': 40500,
    'E55000-75000': 65000,
    'E75000-120000': 97500,
    'E_GE120000': 150000
}).astype(float)

# compute the difference between class_mun_gdp and class
df_nona['diff'] = df_nona['class_mun_gdp'] - df_nona['class_mid']

# check variability in diff
cols = ['diff']

variability(df_nona, cols)
```

```python
# compute the median weighted income for each municipality
def weighted_median(values, weights):
    return wquantiles.median(values, weights)

df_median = df_nona.groupby(['year','mun_istat']).apply(
    lambda g: weighted_median(g['class_mun_gdp'], g['taxpayers'])
).reset_index(name='median_income')
```
